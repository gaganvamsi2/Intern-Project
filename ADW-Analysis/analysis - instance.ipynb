```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
configurations=["1","2","3","4","6","7","1A","2A","3A","4A","6A","7A","100t","200t"]
```

```python
rdata = pd.DataFrame(columns=['configurations', 'errors'])
for i in range(0,len(configurations)):
    file = r'C:\Users\GagaN\Desktop\ADW-Analysis\testingADW\CPU-'
    file=file+configurations[i]+r'\500.csv'
    data= pd.read_csv (file)
    df = pd.DataFrame(data, columns= ['responseCode','success','Latency','Connect'])
    index=df.success.value_counts().index.tolist()
    id = index.index(True)
    errors=500-df.success.value_counts().tolist()[id]
    rdata=rdata.append({'configurations': configurations[i] , 'errors': errors}, ignore_index=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = "configurations", y = "errors", palette = 'Reds',hue='errors', data = rdata,dodge=False).set_title("Service Unavailable errors for different configurations for 500 concurrent users")
#plt.show()
plt.savefig('errors500i.png')
```
